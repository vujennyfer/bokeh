```python
# Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
import geopandas as gpd
from shapely.geometry import Point
from pandas import DataFrame
from geopandas import GeoDataFrame
from bokeh.plotting import save, figure
from bokeh.models import GeoJSONDataSource
```

```python
# Import CSV
# convert the csv file to a DataFrame
data = DataFrame.from_csv('C:/Users/jwhoj/Desktop/Capstone_1/KC_house_data.csv', index_col='zipcode')

# extract the geometry from the DataFrame
points = [Point(row['long'], row['lat']) for key, row in data.iterrows()]

#convert the DataFrame to a GeoDataFrame 
geo_dataframe = GeoDataFrame(data,geometry=points)

# save the resulting shapefile
price = geo_dataframe.to_file('price.shp', driver='ESRI Shapefile') 
```

```python
df = data[['price','bedrooms', 'bathrooms', 'sqft_living', 'waterfront', 'lat', 'long']]
```

```python
# New df w/ variables of interest
df.head()
```

```python
# Make df into geo dataframe 
geo_df = GeoDataFrame(df,geometry=points)
```

```python
# Import King County Shapefile for Zipcode Map 
%matplotlib inline
import geopandas as gpd
kc = gpd.read_file('C:/Users/jwhoj/Desktop/admin/zipcode.shp', index_col='zipcode')
```

```python
points = kc.to_crs({'init': 'epsg:4326'})
```

```python
# Choose variables of focus 
shp = points[['ZIPCODE', 'COUNTY', 'Shape_area', 'Shape_len', 'geometry']]
```

```python
# New dataframe
shp.head()
```

```python
shp.head()
```

```python
shp_map = shp.set_index('zipcode')
```

```python
# Merge CSV and SHPfile
frames = [geo_df, shp_map]
```
